# Сборный проект "Анализ данных о продаже компьютерных игр"
Интернет-магазин «Стримчик» продаёт по всему миру компьютерные игры. Из открытых источников доступны исторические данные о продажах игр, оценки пользователей и экспертов, жанры и платформы (например, Xbox или PlayStation). Нужно выявить определяющие успешность игры закономерности. Это позволит сделать ставку на потенциально популярный продукт и спланировать рекламные кампании.<br>
Имеются данные до 2016 года (данные за сам 2016 год могут быть неполныvb):
- `Name` — название игры;
- `Platform` — платформа;
- `Year_of_Release` — год выпуска;
- `Genre` — жанр игры;
- `NA_sales` — продажи в Северной Америке (миллионы проданных копий);
- `EU_sales` — продажи в Европе (миллионы проданных копий);
- `JP_sales` — продажи в Японии (миллионы проданных копий);
- `Other_sales` — продажи в других странах (миллионы проданных копий);
- `Critic_Score` — оценка критиков (максимум 100);
- `User_Score` — оценка пользователей (максимум 10);
- `Rating` — рейтинг от организации ESRB (англ. Entertainment Software Rating Board). Эта ассоциация определяет рейтинг компьютерных игр и присваивает им подходящую возрастную категорию, например, «Для взрослых», «Для детей младшего возраста» или «Для подростков».

Необходимо провести исследовательский анализ данных, а также проверить следующие гипотезы:
1. Средние пользовательские рейтинги платформ *Xbox One* и *PC* не отличаются.
2. Средние пользовательсике рейтинги жанров *Action* и *Sports* отличаются.

## Шаг 2. Предобработка данных

### Приведение названия столбцов к нижнему регистру

```python
data.columns = data.columns.str.lower()
```

```python
data.head()
```

### Пропущенные значения
Ранее выяснили, что пропуски есть в колонках `name`, `year_of_release`, `genre`, `critic_score`, `user_score`, `rating`. Рассмотрим каждую колонку.

```python
data.isna().sum()
```

```python
pd.DataFrame(round(data.isna().mean().sort_values(ascending=False)*100,1)).style.background_gradient('coolwarm')
```

Видим, что доля пропусков `name`, `genre` и `year_of_release` ничтожно мала, поэтому их можно будет удалить.

#### Пропуски `name`

```python
data['name'].isna().sum()
```

```python
data[data['name'].isna()]
```

В колонке `name` есть два пропуска. Возможно, имела место какая-то техническая ошибка. Восстановить пропущенные названия игр нельзя, поэтому удаляем эти значения.

```python
data = data.drop(data[data['name'].isna()].index).reset_index().drop('index', axis=1)
```

#### Пропуски `year_of_release`

```python
data['year_of_release'].isna().sum()
```

```python
data[data['year_of_release'].isna()].head()
```

В колонке `year_of_release` 269 пропусков. Пропуски могли возникнуть из-за технической ошибки или ошибки внесения данных. Пропуски составляют небольшую долю данных (1.6%), поэтому можно их удалить.

```python
data = data.drop(data[data['year_of_release'].isna()].index).reset_index().drop('index', axis=1)
```

#### Пропуски `genre`

```python
data['genre'].isna().sum()
```

Видимо, после удаления пропусков в предыдущих колонках, мы избавились от пропусков в столбце `genre`.

#### Пропуски `critic_score`

```python
data['critic_score'].isna().sum()
```

В столбце `critic_score` пропущено около половины значений, мы не можем удалить все эти строки. Пропуски могли возникнуть из-за отсутствия информации об оценке критиков.<br> 
Изучим строки с пропущенными значениями.

```python
data[data['critic_score'].isna()].head(10)
```

Мы не можем считать, что данные пропущены совершенно случайно и восстановить значения не представляется возможным. Оставляем пропуски, как есть.

#### Пропуски `user_score`

```python
data['user_score'].isna().sum()
```

В колонке `user_score` нашлось 6633 пропуска. Как и в случае с пропусками в колонке `critic_score`, пропуски в колонке `user_score` могут быть связаны с отсутствием информации о пользовательском рейтинге.<br>
Кроме обычных пропусков в колонке также есть значения `tbd`, которые по сути также являются пропусками. Рассмотрим их подробнее.

```python
data[data['user_score'] == 'tbd'].shape
```

```python
data['user_score'].isna().sum() + data[data['user_score'] == 'tbd'].shape[0]
```

Итого в колонке `user_score` есть 9027 честных пропусков и "псевдопропусков". Преобразуем псевдопропуски в NaN-ы. Воспользуемся функцией `to_numeric`, которая заодно преобразует значения к численному типу, а аргументу `errors` передадим значение `coerce`, чтобы значения, которые нельзя преобразовать (`tbd`), стали пропусками.

```python
data['user_score'] = data['user_score'].apply(pd.to_numeric, errors='coerce', downcast='float')
```

```python
data[data['user_score'] == 'tbd'].shape
```

```python
data['user_score'].isna().sum()
```

Пропуски были приведены к единому виду. Оставляем пропуски незаполненными.

#### Пропуски `rating`

```python
data['rating'].isna().sum()
```

```python
data[data['rating'].isna()].head(20)
```

В столбце `rating` насчитывается 6699 пропусков. Пропуски могли возникнуть из-за того, что рейтинг ERSB разработан для развлекательного программного обеспечения в США и Канаде, т.е. игры, выпущенные в других странах, могут не получить этот рейтинг. Также могла иметь место техническая ошибка (например, видим, что *Call of Duty* не получила рейтинг).

```python
data['rating'] = data['rating'].fillna('Unknown')
```

```python
data['rating'].isna().sum()
```

На предыдущем шаге были преобразованы значения в столбце `user_score` в численные (float). Предлагается также преобразовать значения в колонке `year_of_release` в целочисленные (int). Значения `critic_score` в целочисленные преобразовать не получится из-за того, что в колонке есть NaN-ы.

```python
data.info()
```

```python
data.head()
```

## Шаг 3. Исследовательский анализ данных

### Количество игр, выпущенных в разные годы
Посмотрите, сколько игр выпускалось в разные годы.

## Шаг 5. Проверка гипотез

### Средние пользовательские рейтинги платформ Xbox One и PC одинаковые
Сначала посмотрим на распределения пользовательских рейтингов для этих платформ.

```python
data_actual.query('platform == "XOne"')['user_score'].hist(bins=11)
data_actual.query('platform == "PC"')['user_score'].hist(alpha=0.5,bins=11)
plt.title('Распределения пользовательских рейтингов')
plt.legend(['Xbox One', 'PC'])
plt.xlabel('Оценка пользователя')
plt.ylabel('Частота оценки')
plt.show()
```

```python
data_actual.query('platform == "XOne"')['user_score'].describe()
```

```python
data_actual.query('platform == "PC"')['user_score'].describe()
```

Мы хотим проверить гипотезу о равенстве средних двух генеральных совокупностей.<br>
Нулевая гипотеза $H_0$: пользователи ставят одинаковые оценки играм на Xbox One и на PC.<br>
Альтернативная гипотеза $H_1$: пользователи ставят разные оценки на разных платформах.
Возьмем уровень статистической значимости ${\alpha} = 0.05$.
Для проверки этой гипотезы нам нужно использовать двуcторонний t-критерий Стьюдента.

```python
alpha = 0.05
results_user_score = st.ttest_ind(
    data_actual.query('platform == "XOne"')['user_score'], 
    data_actual.query('platform == "PC"')['user_score'], equal_var=False, nan_policy='omit')

print('Средний пользовательский рейтинг для игр на Xbox One:', data_actual.query('platform == "XOne"')['user_score'].mean())
print('Средний пользовательский рейтинг для игр на PC:', data_actual.query('platform == "PC"')['user_score'].mean())
print('p-значение:', results_user_score.pvalue)


if results_user_score.pvalue < alpha:
    print('Отвергаем нулевую гипотезу')
else:
    print('Не получилось отвергнуть нулевую гипотезу') 
```

Таким образом, мы принимаем нашу нулевую гипотезу: средние пользовательские рейтинги одинаковые для платформ Xbox One и PC.<br>

### Средние пользовательские рейтинги жанров Action и Sports разные
Посмотрим на распределения пользовательских рейтингов для этих жанров.

```python
data_actual.query('genre == "Action"')['user_score'].hist(bins=11)
data_actual.query('genre == "Sports"')['user_score'].hist(alpha=0.5,bins=11)
plt.title('Распределения пользовательских рейтингов')
plt.legend(['Action', 'Sports'])
plt.xlabel('Оценка пользователя')
plt.ylabel('Частота оценки')
plt.show()
```

Распределение оценок для жанра *Sports* более пологое, в отличие от жанра *Action*. На глаз средние значения действительно отличаются.

```python
data_actual.query('genre == "Action"')['user_score'].describe()
```

```python
data_actual.query('genre == "Sports"')['user_score'].describe()
```

Мы хотим проверить гипотезу о равенстве средних двух генеральных совокупностей.<br>
Нулевая гипотеза $H_0$: пользователи ставят одинаковые оценки играм жанра Action и Sports.<br>
Альтернативная гипотеза $H_1$: пользователи ставят разные оценки играм разных жанров.
Возьмем уровень статистической значимости ${\alpha} = 0.05$.<br>
Для проверки этой гипотезы нам нужно использовать двуcторонний t-критерий Стьюдента.

```python
results_user_score_genres = st.ttest_ind(
    data_actual.query('genre == "Action"')['user_score'], 
    data_actual.query('genre == "Sports"')['user_score'], equal_var=False, nan_policy='omit')

print('Средний пользовательский рейтинг для игр в жанре Action:', data_actual.query('genre == "Action"')['user_score'].mean())
print('Средний пользовательский рейтинг для игр в жанре Sports', data_actual.query('genre == "Sports"')['user_score'].mean())
print('p-значение:', results_user_score_genres.pvalue)

if results_user_score_genres.pvalue < alpha:
    print('Отвергаем нулевую гипотезу')
else:
    print('Не получилось отвергнуть нулевую гипотезу') 
```

Мы отвергаем нулевую гипотезу: оценки, которые пользователи ставят играм жанра *Action*, отличаются от оценок игр жанра *Sports*.
