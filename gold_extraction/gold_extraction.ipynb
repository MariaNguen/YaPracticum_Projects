### Проверьте, что эффективность обогащения рассчитана правильно

Напишем функцию `recovery`, которая будет рассчитывать эффективность обогащения.

C — доля золота в концентрате после флотации/очистки;

F — доля золота в сырье/концентрате до флотации/очистки;

T — доля золота в отвальных хвостах после флотации/очистки.

```python
def recovery(c, f, t):
    return (c*(f-t))/(f*(c-t)) * 100
```

```python
C = full_df['rougher.output.concentrate_au']
T = full_df['rougher.output.tail_au']
F = full_df['rougher.input.feed_au']
```

```python
calculated_recovery = recovery(C, F, T)
```

```python
calculated_recovery.head()
```

```python
full_df['rougher.output.recovery'].head()
```

На первый взгляд, значения равны. Рассчитаем местрику MAE.

```python
mae_recovery = mean_absolute_error(full_df['rougher.output.recovery'], calculated_recovery)
```

```python
print(f'MAE = {mae_recovery:.20f}')
```

MAE очень близка к нулю, т.е. мы можем считать, что эффективность обогащения рассчитана верно.

Судя про графикам распределения размера гранцул в сырье различаются не очень сильно, но предлагаю проверить это статистически.

Проверим гипотезу о равенстве средних двух генеральных совокупностей с помощью t-теста.

Нулевая гипотеза $H_0$: размеры гранул в сырье в обучающей и тестовой выборках равны.

Альтернативная гипотеза $H_1$: размеры гранул в сырье в обучающей и тестовой выборках не равны.
Возьмем уровень статистической значимости $\alpha = 0.5$.

```python
alpha = 0.05
```

```python
train_input_feed_size = train_input_feed_size.dropna()
test_input_feed_size = test_input_feed_size.dropna()
```

```python
results_feed_size = st.ttest_ind(train_input_feed_size, test_input_feed_size, 
                                 alternative='two-sided', equal_var=False)

print('Средней размер гранул в сырье в обучающей выборке:', train_input_feed_size.mean())
print('Средней размер гранул в сырье в тестовой выборке:', test_input_feed_size.mean())
print('p-значение:', results_feed_size.pvalue)

if results_feed_size.pvalue < alpha:
    print('Отвергаем нулевую гипотезу')
else:
    print('Не получилось отвергнуть нулевую гипотезу') 
```

Мы не отвергаем гипотезу о равенстве средних.

Можно также проверить гипотезу об одинаковости двух распределений, используя тест Колмогорова-Смирнова.

Нулевая гипотеза $H_0$: размеры гранул в сырье в обучающей и тестовой выборках распределены одинаково.

Альтернативная гипотеза $H_1$: размеры гранул в сырье в обучающей и тестовой выборках рапсределены по-разному.
Возьмем уровень статистической значимости $\alpha = 0.5$.

```python
results_feed_size = st.kstest(train_input_feed_size, test_input_feed_size, 
                                 alternative='two-sided',)

print('p-значение:', results_feed_size.pvalue)

if results_feed_size.pvalue < alpha:
    print('Отвергаем нулевую гипотезу')
else:
    print('Не получилось отвергнуть нулевую гипотезу') 
```

Гипотеза об одинаковости распределений, однако, отвергается на уровне значимости $\alpha = 0.05$.

### Функция для рассчета sMAPE

```python
def total_sMAPE(y_true, y_pred):
    y_true_rougher = np.array(y_true['rougher.output.recovery'])
    y_pred_rougher = np.array(y_pred[:, 0])
    y_true_final = np.array(y_true['final.output.recovery'])
    y_pred_final = np.array(y_pred[:, 1])
    N = len(y_true)
    # np.clip(), чтобы не было деления на 0
    rougher_sMAPE = 100 / N * np.sum(2 * np.abs(y_true_rougher - y_pred_rougher) / np.clip((np.abs(y_true_rougher) + np.abs(y_pred_rougher)), 1e-7, None))
    final_sMAPE = 100 / N * np.sum(2 * np.abs(y_true_final - y_pred_final) / np.clip((np.abs(y_true_final) + np.abs(y_pred_final)), 1e-7, None))
    return 0.25 * rougher_sMAPE + 0.75 * final_sMAPE
```

```python
sMAPE_score = make_scorer(total_sMAPE, greater_is_better = False)
```
